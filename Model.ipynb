###Mapping the data to numbers

```python
frequency_night_map = {
    "Never": 1,
    "Rarely (1-2 times a week)": 2,
    "Sometimes (3-4 times a week)": 3,
    "Often (5-6 times a week)": 4,
    "Every night": 5
}

frequency_day_map = {
    "Never": 1,
    "Rarely (1-2 times a week)": 2,
    "Sometimes (3-4 times a week)": 3,
    "Often (5-6 times a week)": 4,
    "Every day": 5
}

sleep_hours_map = {
    "Less than 4 hours": 1,
    "4-5 hours": 2,
    "6-7 hours": 3,
    "7-8 hours": 4,
    "More than 8 hours": 5
}

stress_level_map = {
    "No stress": 1,
    "Low stress": 2,
    "High stress": 3,
    "Extremely high stress": 4
}

academic_performance_map = {
    "Poor": 1,
    "Below Average": 2,
    "Average": 3,
    "Good": 4,
    "Excellent": 5
}
col7_map = {
    "Very poor":1,
    "Poor":2,
    "Average":3,
    "Good":4,
    "Very good":5

}
sleepdiff_map = {
    "Never": 1,
    "Rarely": 2,
    "Sometimes": 3,
    "Often": 4,
    "Always": 5
}
skip_classes_map = {
    "Never": 1,
    "Rarely (1-2 times a month)": 2,
    "Sometimes (1-2 times a week)": 3,
    "Often (3-4 times a week)": 4,
    "Always": 5
}
sleep_impact_map = {
    "No impact": 1,
    "Minor impact": 2,
    "Moderate impact": 3,
    "Major impact": 4,
    "Significant impact": 4,
    "Severe impact": 5
}
```

###Mapping the values with the map values we have created

```python
df['7. How often do you experience difficulty concentrating during lectures or studying due to lack of sleep?']= \
    df['7. How often do you experience difficulty concentrating during lectures or studying due to lack of sleep?'].map(sleepdiff_map)
df['8. How often do you feel fatigued during the day, affecting your ability to study or attend classes?']= \
    df['8. How often do you feel fatigued during the day, affecting your ability to study or attend classes?'].map(sleepdiff_map)
df["3. How often do you have difficulty falling asleep at night? "] = df["3. How often do you have difficulty falling asleep at night? "].map(frequency_night_map)
df["5. How often do you wake up during the night and have trouble falling back asleep?"] = \
    df["5. How often do you wake up during the night and have trouble falling back asleep?"].map(frequency_night_map)

df["11. How often do you use electronic devices (e.g., phone, computer) before going to sleep?"] = \
    df["11. How often do you use electronic devices (e.g., phone, computer) before going to sleep?"].map(frequency_night_map)

df["12. How often do you consume caffeine (coffee, energy drinks) to stay awake or alert?"] = \
    df["12. How often do you consume caffeine (coffee, energy drinks) to stay awake or alert?"].map(frequency_day_map)

df["13. How often do you engage in physical activity or exercise?"] = \
    df["13. How often do you engage in physical activity or exercise?"].map(frequency_day_map)

df["4. On average, how many hours of sleep do you get on a typical day?"] = \
    df["4. On average, how many hours of sleep do you get on a typical day?"].map(sleep_hours_map)

df["14. How would you describe your stress levels related to academic workload?"] = \
    df["14. How would you describe your stress levels related to academic workload?"].map(stress_level_map)

df["15. How would you rate your overall academic performance (GPA or grades) in the past semester?"] = \
    df["15. How would you rate your overall academic performance (GPA or grades) in the past semester?"].map(academic_performance_map)
df["2. What is your gender?"] = df["2. What is your gender?"].map({
    "Male": 0,
    "Female": 1
})

df["1. What is your year of study?"] = df["1. What is your year of study?"].map({
    "First year": 1,
    "Second year": 2,
    "Third year": 3,
    "Graduate student": 4
})
df["9. How often do you miss or skip classes due to sleep-related issues (e.g., insomnia, feeling tired)?"] = \
    df["9. How often do you miss or skip classes due to sleep-related issues (e.g., insomnia, feeling tired)?"].map(skip_classes_map)
df["10. How would you describe the impact of insufficient sleep on your ability to complete assignments and meet deadlines?"] = \
    df["10. How would you describe the impact of insufficient sleep on your ability to complete assignments and meet deadlines?"].map(sleep_impact_map)

df['6. How would you rate the overall quality of your sleep?']= \
    df['6. How would you rate the overall quality of your sleep?'].map(col7_map)
```

#**Building the Model**

##Splitting Data and training the model

```python
from sklearn.model_selection import train_test_split
x=df_renamed.drop(columns='Academic Performance',axis=1)
y=df_renamed['Academic Performance']
target_log=np.log1p(y)
```

```python
x
```

```python
target_log
```

##Constructing ML Model

```python
x_train,x_test,y_train,y_test=train_test_split(x,target_log,test_size=0.2,random_state=100)
x_train.shape,x_test.shape
```

###**MODEL BUILDING**

####Model: RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
model_forest=RandomForestRegressor(n_estimators=250,max_depth=8,random_state=100)
model_forest.fit(x_train,y_train)
```

```python
y_forest=model_forest.predict(x_test)
```

####Model: XGBoostRegressor

```python
from xgboost import XGBRegressor
model_xg=XGBRegressor(n_estimators=250,max_depth=8,random_state=100,learning_rate=0.25,subsample=0.9)
#sampleweight addition
count=y.value_counts()
weights_dict=(np.sqrt(len(df_renamed))/count).to_dict()
samp_weights=y.map(weights_dict)
model_xg.fit(x_train,y_train,sample_weight=samp_weights[y_train.index])
```

```python
y_xgb=model_xg.predict(x_test)
```

##Checking efficiency of Model

###RandomForestRegressor *Evaluation*

```python
#Random Forest
from sklearn.metrics import r2_score,mean_absolute_error
model_r2=r2_score(y_test,y_forest)
model_mae=mean_absolute_error(y_test,y_forest)
print(f"R2 Score:{model_r2}")
print(f"Mean Absolute Error:{model_mae}")
```

```python
errors = y_forest - y_test
plt.figure(figsize=(8, 5))
plt.hist(errors, bins=20)
plt.title('Distribution of Prediction Errors')
plt.xlabel('Prediction Error')
plt.ylabel('Frequency')
plt.show()
```

```python
plot_df=pd.DataFrame({'Actual Values':y_test,'Predicted Values':y_forest})
sns.boxplot(
    x="Actual Values",
    y="Predicted Values",
    data=plot_df
)
```

###XGBBoostRegressor *evaluation*

```python
from sklearn.metrics import r2_score,mean_absolute_error
model_r2=r2_score(y_test,y_xgb)
model_mae=mean_absolute_error(y_test,y_xgb)
print(f"R2 Score:{model_r2}")
print(f"Mean Absolute Error:{model_mae}")
```

```python
errors = y_xgb - y_test
plt.figure(figsize=(8, 5))
plt.hist(errors, bins=30)
plt.title('Distribution of Prediction Errors')
plt.xlabel('Prediction Error')
plt.ylabel('Frequency')
plt.show()
```

```python
plot_df=pd.DataFrame({'Actual Values':y_test,'Predicted Values':y_xgb})
sns.boxplot(
    x="Actual Values",
    y="Predicted Values",
    data=plot_df
)
```
